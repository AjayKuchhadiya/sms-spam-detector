## Data Preprcessing

```python
# function to clean html tags
import re
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['text']= df['text'].apply(clean_html)
```

```python
df.head()
```

```python
# converting everything to lower case

def convert_lower(text):
    return text.lower()
```

```python
df['text'] = df['text'].apply(convert_lower)
```

```python
df.head()
```

```python
# function to remove special characters

def remove_special(text):
    x = ''
    
    for i in text :
        if i.isalnum():     # asking evry word if you are alpha-numeric or not -- if not the replace with empty space
            x = x+i           # if yes then concatenate with empty string
        else :
            x = x + ' '
    return x
```

```python
df['text'] = df['text'].apply(remove_special)
```

```python
df.head()
```

```python

# to remove stop words
import nltk
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y = x[:]
    x.clear()
    return y
        
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```

```python
df.head()
```

```python
# perform stemming -- brings the words in their root form
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
y = []
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z = y[:]
    y.clear()
    return z
```

```python
stem_words(['i', 'loved', 'loving', 'it'])
```

```python
df['text'] = df['text'].apply(stem_words)
```

```python
df.head()
```

```python
# Join back

def join_back(list_input):
    return ' '.join(list_input)
```

```python
df['text'] = df['text'].apply(join_back)
```

## Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = CountVectorizer(max_features =2000)
tf = TfidfVectorizer(max_features =2000)
```

```python
X = cv.fit_transform(df['text']).toarray()
```

```python
# X = tf.fit_transform(df['text']).toarray()
```

```python
X.shape
```

```python
y = df['target'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state  = 5
                                                  )
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_train, y_train)
mnb.fit(X_train, y_train)
bnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
y_pred_mnb = mnb.predict(X_test)
y_pred_bnb = bnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
print('accuracy score of gnb = ', accuracy_score(y_pred_gnb, y_test))
print('accuracy score of mnb = ', accuracy_score(y_pred_mnb, y_test))
print('accuracy score of bnb = ', accuracy_score(y_pred_bnb, y_test))
```

```python
print('confusion mat of gnb = ', confusion_matrix(y_pred_gnb, y_test))
print('confusion mat of mnb = ', confusion_matrix(y_pred_mnb, y_test))
print('confusion mat of bnb = ', confusion_matrix(y_pred_bnb, y_test))
```

```python
print('precision score of gnb = ', precision_score(y_pred_gnb, y_test))
print('precision score of mnb = ', precision_score(y_pred_mnb, y_test))
print('precision score of bnb = ', precision_score(y_pred_bnb, y_test))
```

Here we can see that MultinomialNB is giving us the best result
<br>Precision value matters more here.
